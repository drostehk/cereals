```python
from PIL import Image
import pandas as pd
```

```python
df = pd.read_csv('../data/data_updated.csv')
df = df.drop('Unnamed: 0', 1)
df.head()
```

```python
cropped_img_name = df['file_name_cropped']
print type(cropped_img_name[0])
print cropped_img_name.head()
print len(cropped_img_name)
```

```python
# Most frequent colour in image (background excluded)
def most_frequent_colour(image):
    w, h = image.size
    pixels = image.getcolors(w * h)
    
    most_frequent_pixel = pixels[0]
    
    for count, colour in pixels:
        if (count > most_frequent_pixel[0] & colour[3] != 0):
            most_frequent_pixel = (count, colour)
            
    return most_frequent_pixel
```

```python
print cropped_img_name[0]
```

```python
# Generate the most frequent colour feature with its count

counts = []
hexcodes = []

for i in range(260):
    img_name = '../img/kelvin_cropped/' + cropped_img_name[i]
    img = Image.open(img_name)
    count, fre_colour =  most_frequent_colour(img)
    rgb = fre_colour[:-1]
    hexcode = '#'+''.join(map(chr, rgb)).encode('hex')
    print str(i + 1).zfill(3) + ': ' + hexcode + ' with count: ' + str(count)
    
    counts.append(count)
    hexcodes.append(hexcode)
```

```python
print counts[:5]
```

```python
print hexcodes[:5]
```

```python
def genFreColor(row):
    return hexcodes[row['index'] - 1]

def genFreColorCount(row):
    return counts[row['index'] - 1]
```

```python
df['most_fre_color'] = df.apply(genFreColor, axis = 1)
df['most_fre_color_count'] = df.apply(genFreColorCount, axis = 1)
```
